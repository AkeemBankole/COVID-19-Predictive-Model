Data Cleaning: This involves merging datasets, handling missing values, remove duplicates, data transformation and standardize date and location formats.

```python
# Merge files together
# Define file names
files_to_merge = ["country_wise_latest.csv", "day_wise.csv", "full_grouped.csv", "worldometer_data.csv"]
```

```python
# Load the files into pandas DataFrames
dfs = {file: pd.read_csv(os.path.join(extract_path, file)) for file in files_to_merge}

```

```python
# Display initial info about each file
for name, df in dfs.items():
    print(f"\n{name} - Shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")

```

```python
# Perform merging
# Example: Merge based on a common column like 'Country/Region' or 'Date'
merged_df = dfs["country_wise_latest.csv"]
merged_df = merged_df.merge(dfs["full_grouped.csv"], on="Country/Region", how="inner")
merged_df = merged_df.merge(dfs["worldometer_data.csv"], on="Country/Region", how="inner")
merged_df = merged_df.merge(dfs["day_wise.csv"], on="Date", how="inner")

```

```python
# Display information about the merged DataFrame
print("\nMerged DataFrame:")
print(f"Shape: {merged_df.shape}")
print(f"Columns: {merged_df.columns.tolist()}")
print(merged_df.head())
```

```python
# Check for duplicates in the merged_df
duplicate_rows = merged_df[merged_df.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
```

```python
# Check for missing values in the merged_df
missing_values = merged_df.isnull().sum()
print(missing_values)
```

```python
# Handle missing values in the merged_df
# Fill 'NewCases', 'NewDeaths', and 'NewRecovered' with 0 (missing likely means no cases)
columns_fill_zero = ['NewCases', 'NewDeaths', 'NewRecovered']
merged_df[columns_fill_zero] = merged_df[columns_fill_zero].fillna(0)

```

```python
# Fill columns like 'TotalDeaths', 'TotalRecovered', 'ActiveCases', 'Serious,Critical' with median values
columns_fill_median = ['TotalDeaths', 'TotalRecovered', 'ActiveCases', 'Serious,Critical']
for column in columns_fill_median:
    merged_df[column] = merged_df[column].fillna(merged_df[column].median())
```

```python
# Fill 'Deaths/1M pop', 'TotalTests', 'Tests/1M pop' with mean values
columns_fill_mean = ['Deaths/1M pop', 'TotalTests', 'Tests/1M pop']
for column in columns_fill_mean:
    merged_df[column] = merged_df[column].fillna(merged_df[column].mean())

```

```python
# Verify all missing values are handled
print("\nMissing Values After Handling:")
print(merged_df.isnull().sum())
```

```python
# Standardize the Date column
if 'Date' in merged_df.columns:
    merged_df['Date'] = pd.to_datetime(merged_df['Date'], errors='coerce')  # Convert to datetime
    print("\nDate column standardized to datetime format.")
    print(merged_df['Date'].head())  # Check the first few rows of the Date column

```

```python
# Standardize location names (Country/Region)
if 'Country/Region' in merged_df.columns:
    # Example: Replace inconsistent country names with standardized ones
    country_replacements = {
        'United States': 'USA',
        'United Kingdom': 'UK',
        'South Korea': 'Korea, South',
        'Russia': 'Russian Federation',
        # Add more replacements as needed
    }
    merged_df['Country/Region'] = merged_df['Country/Region'].replace(country_replacements)
    print("\nCountry/Region column standardized.")
    print(merged_df['Country/Region'].unique())  # Display unique values for verification
```

```python
# Check for numerical columns in the merged_df
numerical_columns = merged_df.select_dtypes(include=np.number).columns
```

```python
# Check for infinite values
print("\nChecking for infinite values in merged_df:")
print(merged_df[numerical_columns].isin([float('inf'), float('-inf')]).sum())

# Check for extremely large values
print("\nDescriptive statistics for numerical columns:")
print(merged_df[numerical_columns].describe())
```

```python
# Replace infinite values with NaN
merged_df[numerical_columns] = merged_df[numerical_columns].replace([float('inf'), float('-inf')], pd.NA)

# Check if there are still any NaN values after replacing infinities
print("\nChecking for NaN values after replacing infinities:")
print(merged_df.isna().sum())
```

```python
# Fill NaN values with the mean or median of the column
merged_df[numerical_columns] = merged_df[numerical_columns].fillna(merged_df[numerical_columns].mean())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply normalization
merged_df[numerical_columns] = scaler.fit_transform(merged_df[numerical_columns])

print("\nData Normalization Complete:")
print(merged_df[numerical_columns].describe())
```

Feature Engineering: Create derived variables, such as case rates per
population or case increase ratios, to potentially enhance model insights.

```python
# Ensure 'Population' column exists and has no zeros to avoid division errors
if 'Population' in merged_df.columns:
    merged_df['CaseRatePerPopulation'] = merged_df['TotalCases'] / merged_df['Population']
    merged_df['CaseRatePerPopulation'] = merged_df['CaseRatePerPopulation'].fillna(0)  # Handle missing values
else:
    print("Error: 'Population' column not found in the dataset.")
```

```python
# Calculate Case Increase Ratio
if 'NewCases' in merged_df.columns and 'TotalCases' in merged_df.columns:
    merged_df['CaseIncreaseRatio'] = merged_df['NewCases'] / merged_df['TotalCases'].shift(1)
    merged_df['CaseIncreaseRatio'] = merged_df['CaseIncreaseRatio'].fillna(0)  # Handle missing values
else:
    print("Error: Required columns for 'Case Increase Ratio' not found in the dataset.")
```

```python
# Tests per Population
if 'TotalTests' in merged_df.columns and 'Population' in merged_df.columns:
    merged_df['TestsPerPopulation'] = merged_df['TotalTests'] / merged_df['Population']
    merged_df['TestsPerPopulation'] = merged_df['TestsPerPopulation'].fillna(0)
```

```python
# Mortality Rate per Population
if 'TotalDeaths' in merged_df.columns and 'Population' in merged_df.columns:
    merged_df['MortalityRatePerPopulation'] = merged_df['TotalDeaths'] / merged_df['Population']
    merged_df['MortalityRatePerPopulation'] = merged_df['MortalityRatePerPopulation'].fillna(0)
```

```python
print("\nNewly Engineered Features:")
print(merged_df[['CaseRatePerPopulation', 'CaseIncreaseRatio', 'TestsPerPopulation', 'MortalityRatePerPopulation']].head())
```

```python
# Ensure 'Date' is in datetime format
merged_df['Date'] = pd.to_datetime(merged_df['Date'], errors='coerce')

# Check if there are any invalid dates
print(merged_df['Date'].isnull().sum())
```

```python
# Calculate Case Rate and Mortality Rate Per Population
merged_df['CaseRatePerPopulation'] = merged_df['TotalCases'] / merged_df['Population']
merged_df['MortalityRatePerPopulation'] = merged_df['TotalDeaths'] / merged_df['Population']
```

```python
# Handle any infinite or missing values (e.g., division by zero)
merged_df.replace([float('inf'), -float('inf')], float('nan'), inplace=True)
merged_df.dropna(subset=['CaseRatePerPopulation', 'MortalityRatePerPopulation'], inplace=True)
```

```python
# Group by 'Date' and calculate mean Case and Mortality Rates
case_rate_time = merged_df.groupby('Date')['CaseRatePerPopulation'].mean().reset_index()
mortality_rate_time = merged_df.groupby('Date')['MortalityRatePerPopulation'].mean().reset_index()
```
